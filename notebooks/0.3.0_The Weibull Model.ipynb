## 0.3.3 Compare Weibull model parameters

```python
parole = (prison['paro'] == 1)
```

```python
wb.fit(durations=prison[parole]['week'], event_observed=prison[parole]['arrest'])

# Print rho
print("The rho_ parameter of parolee survival function is: ", wb.rho_)
```

```python
# Fit to non parolee
wb.fit(durations=prison[~parole]['week'], event_observed=prison[~parole]['arrest'])

# Print rho
print("The rho_ parameter of non-parolee survival function is: ", wb.rho_)
```

**Question** 
You suspect that parolees' arrest rate decreases as time goes on while other convicts' arrest rate increases as time goes on.

Based on the  values printed and what you know about the Weibull distribution, is your suspicion verified?

R:/ When rho or k is larger than 1, the rate of event increases as time goes on. No, both populations' rate of event increases as time goes on.
Both rho values are larger than 1, which indicates that both populations' rate of event increases as time goes on.

## 0.3.12 Choose a parametric model

```python
heart_patients.head()
```

```python
from lifelines import WeibullFitter, ExponentialFitter, LogNormalFitter
```

```python
wb = WeibullFitter()
exp = ExponentialFitter()
log = LogNormalFitter()
```

```python
for model in [wb, exp, log]:
    model.fit(durations=heart_patients['survival'], event_observed=heart_patients['observed'])
    print(model.__class__.__name__, model.AIC_)
```

```python
print(wb.AIC_, exp.AIC_, log.AIC_)
```

**Question**
Among the 3 models you assessed, which one fits the data the best based on AIC?

R/: WeibullFitter, because is the lowest. <br>
The Weibull model has the lowest AIC value, which means it is the best choice for the data out of all 3.
